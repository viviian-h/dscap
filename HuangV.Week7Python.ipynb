```python
st = acts2024[acts2024['TESCHENR'] == 1]
s = st[(st['TEAGE'] > 14) | (st['TEAGE'] < 25)]

g = s[s['TESEX'] == 2]
b = s[s['TESEX'] == 1]
```

```python
# get total amount of time spent doing each educational activity by gender
g['starttime'] = pd.to_datetime(g['TUSTARTTIM'], format='%H:%M:%S')
g['stoptime'] = pd.to_datetime(g['TUSTOPTIME'], format='%H:%M:%S')
g['len_time'] = g['stoptime'] - g['starttime']

b['starttime'] = pd.to_datetime(b['TUSTARTTIM'], format='%H:%M:%S')
b['stoptime'] = pd.to_datetime(b['TUSTOPTIME'], format='%H:%M:%S')
b['len_time'] = b['stoptime'] - b['starttime']

b['len_time_min'] = b['len_time'].dt.total_seconds() / 60
g['len_time_min'] = g['len_time'].dt.total_seconds() / 60
```

```python
# get all educational activites (starts with 06 in TRCODE) then disaggregate by second tier
ed_g = g[(g['TRCODE'] > 59999) & (g['TRCODE'] < 70000)]['len_time_min'].sum()
ed_b = b[(b['TRCODE'] > 59999) & (b['TRCODE'] < 70000)]['len_time_min'].sum()
g

# total number of girls and guys
g_total = g['TUCASEID'].nunique()
b_total = b['TUCASEID'].nunique()

# Get each individual education break down 
print(ed_g/g_total) # average number of minutes each girl spends
print(ed_b/b_total) # average number of minutes each boy spends

# percent of total time on x educational activity
# class
classdeg_g = g[(g['TRCODE'] == 60101)]['len_time_min'].sum()/g_total
classdeg_b = b[(b['TRCODE'] == 60101)]['len_time_min'].sum()/b_total

classoth_g = g[(g['TRCODE'] >= 60102) & (g['TRCODE'] < 60200)]['len_time_min'].sum()/g_total
classoth_b = b[(b['TRCODE'] >= 60102) & (b['TRCODE'] < 60200)]['len_time_min'].sum()/b_total

# extracurriculars
exc_g = g[(g['TRCODE'] > 60199) & (g['TRCODE'] < 60300)]['len_time_min'].sum()/g_total
exc_b = b[(b['TRCODE'] > 60199) & (b['TRCODE'] < 60300)]['len_time_min'].sum()/b_total

# Research/homework
hw_g = g[(g['TRCODE'] > 60299) & (g['TRCODE'] < 60400)]['len_time_min'].sum()/g_total
hw_b = b[(b['TRCODE'] > 60299) & (b['TRCODE'] < 60400)]['len_time_min'].sum()/b_total

# Registration/admin activities
reg_g = g[(g['TRCODE'] >= 60400) & (g['TRCODE'] < 70000)]['len_time_min'].sum()/g_total
reg_b = b[(b['TRCODE'] >= 60400) & (b['TRCODE'] < 70000)]['len_time_min'].sum()/b_total
```

```python
# 2 stacked bars with percent time spent on educational activity by gender
# then each stacked thing is a different educational activity
classdeg_g
```

```python

data = {
    'Class for Degree':  {'Male': classdeg_b, 'Female': classdeg_g},
    'Other Class':       {'Male': classoth_b, 'Female': classoth_g},
    'Extracurricular':   {'Male': exc_b,      'Female': exc_g},
    'Homework':          {'Male': hw_b,       'Female': hw_g},
    'Administrative':           {'Male': reg_b,      'Female': reg_g},
}

pct_df = pd.DataFrame(data).T

fig = go.Figure()

for activity in pct_df.index:
    fig.add_trace(go.Bar(
        name=activity,
        x=['Male', 'Female'],
        y=[pct_df.loc[activity, 'Male'], pct_df.loc[activity, 'Female']],
        hovertemplate='<b>%{x}</b><br>' + activity + '<br>Avg Minutes: %{y:.1f} min<extra></extra>'
    ))

fig.update_layout(
    barmode='stack',
    title='Average Minutes per Day Spent on Educational Activities by Gender in 2024',
    yaxis_title='Average Minutes per Day',
    xaxis_title='Gender',
    legend_title='Activity',
    hovermode='x'
)

fig.show()
fig.write_html("barplot2024.html")
```
